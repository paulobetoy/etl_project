```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
approval_load = "./Sources/Approval_list.csv"
header_load = "./Sources/Order_header_list.csv"
line_load = "./Sources/Order_line_list.csv"
star_load = "./Sources/STARS_classification_2020.csv"
rate_load = "./Sources/EX_Rate_Apr2020.csv"

approval_data = pd.read_csv(approval_load)
header_data = pd.read_csv(header_load)
line_data = pd.read_csv(line_load)
star_data = pd.read_csv(star_load)
rate_data = pd.read_csv(rate_load)

print(rate_data.columns.values)
print(star_data.columns.values)
print(header_data.columns.values)
print(approval_data.columns.values)
print(line_data.columns.values)
```

```python
currency = rate_data["Currency"]
concatenate = rate_data["concatenate"]
exchange_rate = round(rate_data["ex rate to use 2019"],6)

rate_df = pd.DataFrame({"Currency":currency,
                        "Exchange Code":concatenate,
                        "Exchange Rate to EUR":exchange_rate})

rate_df.head()
```

```python
# Export DataFrame to CSV File
export_csv = rate_df.to_csv ('./Output_Files/ex_rate_modified.csv', index = None, header=True)
```

```python
market = star_data["Market/ Marché"]
category = star_data["Category / Categorie"]
technology = star_data["Technological Group / Regroupement Technologique"]
family = star_data["Family / Famille"]

star_df = pd.DataFrame({"Family":family,
                        "Market":market,
                        "Category":category,
                        "Technology":technology})

star_df.head()
```

```python
header_df = header_data.drop(columns=['Items', 'Total','Currency','Unanswered Comments'])
header_df.head()
```

```python
req_num = header_df["Req"]
PO_num = header_df["PO Number"]
order_date = header_df["Order Date"]
status = header_df["Status"]
t_status = header_df["Transmission Status"]

header_df_mod = pd.DataFrame({"PO Number":PO_num,
                              "Requisition Number":req_num,
                              "Order Date":order_date,
                              "Status":status,
                              "Transmission Method":t_status})

header_df_mod.head()
```

```python
header_df_mod.drop_duplicates("Requisition Number", inplace=True)
header_df_mod.head()
```

```python
# Export DataFrame to CSV File
export_csv = header_df_mod.to_csv ('./Output_Files/header_modified.csv', index = None, header=True)
```

```python
line_data["PO-ID"] = str(line_data["PO Number (Header)"]) + "-" + str(line_data["Line"])
line_data["PO-ID"]
```

```python
line = line_data["Line"]
po_num = line_data["PO Number (Header)"]
order_status = line_data["Order Status (Header)"]
t_status = line_data["Order Transmission Status (Header)"]
item = line_data["Item"]
line_total = line_data["Line Total"]
currency = line_data["Currency"]
account = line_data["Account"]
chart = line_data["Chart Of Accounts"]
contract = line_data["Contract Number"]
erp = line_data["ERP"]
item_class = line_data["Item Classification"]
account_total = line_data["Accounting Total Currency"]



line_df = pd.DataFrame({"PO Number":po_num,
                        "PO Line Number":line,
                        "Item Classification":item_class,
                        "Item Description":item,
                        "Line Total":line_total,
                        "Currency":currency,
                        "Order Status":order_status,
                        "Order Transmission Status":t_status,
                        "Chart Of Accounts":chart,
                        "Account":account,
                        "Account Total Currency":account_total,
                        "Contract Number":contract,
                        "ERP":erp})

line_df.head()
```

```python
line_df.count()
```

```python
line_df.dtypes
```

```python
# Export DataFrame to CSV File
export_csv = line_df.to_csv ('./Output_Files/line_modified.csv', index = None, header=True)
```
